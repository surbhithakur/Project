```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n = 516
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X)
plt.plot(X,Y)
plt.title('Line Graph')
plt.xlabel('X ENTRIES')
plt.ylabel('Y ENTRIES')
plt.savefig('Line Graph.png')
```

```python
data=pd.read_csv('WeightLoss.csv')
data.head()
```

```python
data.plot( figsize=(10, 10),x="Unnamed: 0", y=["wl1", "wl2", "wl3","se1","se2","se3"], kind="barh")
plt.title('Weight Loss Graph')
plt.xlabel('Unnamed:0')
plt.ylabel('WeightLoss for wl1,wl2,wl3,se1,se2,se3')
plt.savefig('WeightLoss for wl1,wl2,wl3,se1,se2,se3.png')
```

```python
data.plot( figsize=(15,15),x="Unnamed: 0", y=["se3"], kind="bar")
plt.title('Weight Loss Graph')
plt.xlabel('Unnamed:0')
plt.ylabel('WeightLoss for se3')
plt.savefig('WeightLoss for se3.png')
```

```python
import seaborn as sns
g=sns.pairplot(data,diag_kind='kde',plot_kws={'alpha':0.2})
g.fig.suptitle("WEIGHT LOSS GRAPH", y=1.08)
g.savefig("output.png")
```
